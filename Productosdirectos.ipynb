<div class='def'>
<strong>$\color{ForestGreen}{\bf Definición}$</strong> <em>[Producto Cartesiano
    ]</em>
<p> El $\color{red}{\bf producto}$ $\color{red}{\bf cartesiano}$ $\color{red}{\bf de}$ 
    $\color{red}{\bf conjuntos}$ $S_1,S_2,\dots ,S_n$ es el conjunto de todas las $n$-ádas odenadas $(a_1,a_2,\dots,a_n)$, donde $a_i\in S_i$. El producto cartesiano se denota por
    $$S_1\times S_2\times\cdots\times S_n$$
    o por
    $$\prod_{i=1}^n S_i$$
    
 </p>
</div> 

<div class='teo'>
<strong>$\color{Maroon}{\bf Teorema}$</strong> 
<p> Sean los grupos $G_1,G_2\dots,G_n$. Para $(a_1,a_2,\dots,a_n)$, $(b_1,b_2,\dots,b_n)\in 
    \prod_{i=1}^n G_i$, defínase $(a_1,a_2,\dots,a_n)(b_1,b_2,\dots,b_n)$$=
    (a_1b_1,a_2b_2,\dots,a_nb_n)$. Entonces $\prod_{i=1}^nG_i$ es un grupo, el $\color{red}{\bf producto}$ $\color{red}{\bf directo}$ $\color{red}{\bf externo}$ $\color{red}{\bf de}$ $\color{red}{\bf los}$ $\color{red}{\bf grupos}$ $G_i$, bajo la operación binaria.
</p>
</div>

<div class='teo'>
<strong>$\color{Maroon}{\bf Teorema}$</strong> 
<p> El grupo ${\mathbb Z}_m\times{\mathbb Z}_n$ es isomorfo a ${\mathbb Z}_{mn}$ si y sólo si $m$ y
    $n$ son primos relativos, esto es, si el $mcd$ de $m$ y $n$ es 1.
</p>
</div>

[Demostración](https://jamboard.google.com/d/1Lbgpf-14Zxe-jtdIqegFJyIfQpc7LbilaaxdTAYU3eE/viewer?f=7)

<div class='teo'>
<strong>$\color{Maroon}{\bf Corolario}$</strong> 
<p> El grupo $\prod_{i=1}^n {\mathbb Z}_{m_i}$ es cíclico e isomorfo a ${\mathbb Z}_{m_1m_2\dots m_n}$ 
    si y sólo si mcd($m_i,m_j$)=1 para $i\neq j$.
</p>
</div>

El corolario anterior muestra que si $n$ se escribe como 
$$n=p_1^{n_1}p_2^{n_2}\dots np_r^{n_r},$$ 
con $p_i$ primos distintos, entonces
$${\mathbb Z}_n\simeq {\mathbb Z}_{p_1^{m_1}}\times {\mathbb Z}_{p_2^{m_2}}\times
\dots \times {\mathbb Z}_{p_r^{m_r}}.$$
En particular, ${\mathbb Z}_{72}\simeq {\mathbb Z}_8\times{\mathbb Z}_9$

<div class='def'>
<strong>$\color{ForestGreen}{\bf Definición}$</strong> <em>[Orden de un grupo]</em>
<p> Sea $G$ un grupo y $a\in G$. Si existe algún entero positivo $n$ tal que $a^n=e$,
    el menor de dichos enteros positivos $n$, es el $\color{red}{\bf orden}$
    $\color{red}{\bf de}$ $a$. Si no existe dicha $n$, entonces $a$ es de $\color{red}{\bf orden}$
    $\color{red}{\bf infinito}$.   
 </p>
</div> 

De aquí que *si $a$ es un elemento de un grupo $G$, el orden de $a$ es igual al orden del subgrupo
cíclico generado pra $a$*. 

<div class='teo'>
<strong>$\color{Maroon}{\bf Teorema}$</strong> 
<p> Sea $(a_1,a_2,\dots ,a_n)\in \prod_{i=1}^n G_i$. Si $a_i$ es de orden finito $r_i$ en $G_i$, 
    entonces el orden de $(a_1,a_2,\dots ,a_n)$ en $\prod_{i=1}^n G_i$ es igual al mínimo común múltiplo
    de todas las $r_i$.
</p>
</div>

<div class='def'>
<strong>$\color{ForestGreen}{\bf Definición}$</strong> <em>[Producto Interno
    ]</em>
<p> Sea un grupo $G$ con subgrupos $H_i$ para $i=1,2,\dots ,n$. $G$ es el
    $\color{red}{\bf producto}$ $\color{red}{\bf directo}$ $\color{red}{\bf interno}$ 
    $\color{red}{\bf de}$ $\color{red}{\bf los}$ $\color{red}{\bf subgrupos}$ $H_i$ si la transformación     $$\phi:\prod_{i=1}^n H_i\rightarrow G$$
    dada por
    $$\phi(h_1,h_2,\dots,h_n)=h_1h_2\dots h_n$$
    es un isomorfismo.
 </p>
</div> 

Es claro que si $\prod_{i=1}^nG_i$ es un producto directo externo de grupos $G_i$, el subconjunto
$$\overline{G}=\{(e_1,e_2,\dots,a_i,e_{i+1},\dots,e_n)\,|\,a_i\in G\},$$
esto es, el conjunto de todas las $n-$ádas con los elementos identidad en todos los lugares excepto el
$i-$ésimo, es un subgrupo de $\prod_{i=1}^nG_i$. Claramente este subgrupo $\overline{G}_i$ es naturalmente isomorfo a $G_i$ bajo la correspondencia dada por la proyección $\pi_i$, donde
$$\pi_i(e_1,e_2,\dots,a_i,e_{i+1},\dots,e_n)=a_i.$$
El grupo $G_i$ se refleja en la $i-$ésima componente de los elementos de $\overline{G}_i$, y las $e_j$ en las otras componentes simplemente van de acompañantes. Consideremos el *producto directo interno* de estos subgrupos $\overline{G}_i$. Los términos *interno* y *externo*, aplicados a los productos directos de grupos, sólo reflejan si se consideran o no (respectivamente), a los grupos componentes como subgrupos del grupo producto.

<div class='teo'>
<strong>$\color{Maroon}{\bf Teorema}$</strong> 
<p> Si $G$ es el producto directo interno de los subgrupos $H_1,H_2,\dots,H_n$, entonces cada $g\in G$ puede escribirse de manera única como $g=h_1h_2\dots h_n$, donde $h_i\in H_i$.
</p>
</div>

<div class='def'>
<strong>$\color{ForestGreen}{\bf Definición}$</strong> <em>[Ensamble
    ]</em>
<p> Sean $H$ y $K$ subgrupos de un grupo  $G$. El $\color{red}{\bf ensamble}$ $H\vee K$ de $H$ y $K$ es
    la intrsección de todos los subgrupos de $G$ que contienen $HK=\{hk\,|\, h\in H,\; k\in K\}$.
 </p>
</div> 

<div class='teo'>
<strong>$\color{Maroon}{\bf Teorema}$</strong> 
<p> Un grupo $G$ es el producto directo interno de subgrupos $H$ y $K$ si y sólo si
    1.- $G=H\vee K$ $\qquad\qquad$
    2.- $hk=kh$ para todas $h\in H$ y $k\in K$$\qquad\qquad$
    3.- $H\cap K=\{e\}$.
</p>
</div>

# Generadores

<div class='def'>
<strong>$\color{ForestGreen}{\bf Definición}$</strong> <em>[Generadores
    ]</em>
<p> Sea $G$ un grupo y $a_i\in G$ para $i\in I$. El menor subgrupo de $G$ que contiene 
    $\{a_i\,|\, i\in I\}$ es el $\color{red}{\bf subgrupo}$ $\color{red}{\bf generado}$ 
    $\color{red}{\bf por}$ $\{a_i\,|\, i\in I\}$. Si este subgrupo es todo $G$, entonces
    $\{a_i\,|\, i\in I\}$ $\color{red}{\bf genera}$ a $G$ y las $a_i$ son $\color{red}{\bf generadores}$
    de $G$. Si existe un conjunto finito $\{a_i\,|\, i\in I\}$ que genere $G$, entonces $G$ es
    $\color{red}{\bf finitamente}$ $\color{red}{\bf generado}$.
 </p>
</div>

<div class='teo'>
<strong>$\color{Maroon}{\bf Teorema}$</strong> 
<p> Si $G$ es un grupo y $a_i\in G$ para $i\in G$, entonces el subgrupo $H$ de $G$ generado por 
    $\{a_i\, | \, i\in I\}$ consta de precisamente aquellos elementos de $G$ que son productos finitos de potencias de exponente entero de $a_i$, donde, en ese producto, pueden presentarse varias veces potencias de alguna $a_i$ dada.
</p>
</div>

<div class='def'>
<strong>$\color{ForestGreen}{\bf Definición}$</strong> <em>[Grupo de torsión
    ]</em>
<p> Un grupo $G$ es un $\color{red}{\bf grupo}$ $\color{red}{\bf de}$ $\color{red}{\bf torsión}$
    si todo elemento de $G$ es de orden finito. $G$ es $\color{red}{\bf libre}$ $\color{red}{\bf de}$
    $\color{red}{\bf torsión}$ si ningún elemento, aparte de la identidad, es de orden finito. 
 </p>
</div>

<div class='teo'>
<strong>$\color{Maroon}{\bf Teorema}$</strong> 
<p> En un grupo abeliano $G$ el conjunto $T$ de tofdos los elementos de $G$ de orden finito es un 
    subgrupo de $G$, el $\color{red}{\bf subgrupo}$ $\color{red}{\bf de}$ $\color{red}{\bf torsión}$
    de $G$.
</p>
</div>

<strong>$\color{Maroon}{\bf Lema}$</strong> 
<p> Si $G$ es un grupo abeliano finitamente generado con subgrupo de torsión $T$, 
    entonces $G$ es un producto directo (interno)
    $T\times F$ para algún subgrupo $F$ de $G$ que sea libre de torsión.
    
</p>
</div>

<div class='teo'>
<strong>$\color{Maroon}{\bf Lema}$</strong> 
<p> Un grupo abeliano $F$ finitamente generado, libre de torsión, es isomorfo a 
    $${\mathbb Z}\times{\mathbb Z}\times\dots\times{\mathbb Z},$$ para algún número $m$ de factores. El número $m$, el número de betti de $F$, es único.   
</p>
</div>

<div class='teo'>
<strong>$\color{Maroon}{\bf Lema}$</strong> 
<p> En un grupo abeliano finito $T$ es isomorfo a dos tipos diferentes de productos directos de grupos cíclicos, como sigue
    
</p>
</div>


1) $T$ es isomorfo a un producto 
$$T\simeq {\mathbb Z}_{p_1^{r_1}}\times{\mathbb Z}_{p_2^{r_2}}\times\dots\times{\mathbb Z}_{p_s^{r_s}}$$
donde los $p_i$ son números primos, no necesariamente distintos. Este producto directo de grupos cíclicos de orden la potencia de un número primo isomorfo a $T$, es único excepto por un arreglo de los factores.

2) $T$ es isomorfo a un producto directo
$$T\simeq{\mathbb Z}_{m_1}\times{\mathbb Z}_{m_2}\times\dots\times{\mathbb Z}_{m_r},$$
donde $m_i$ divide a $m_{i+1}$. Los números $m_i$, los coeficientes de torsión de $T$ son únicos.
